```python
def resize_and_padding(image , target_size):
    image.thumbnail(target_size , Image.LANCZOS)
    new_image = Image.new("RGB" , target_size ,  (255, 255, 255))
    left = (target_size[0] - image.width) // 2
    top  = (target_size[1] - image.height) // 2
    
    new_image.paste(image , (left , top))
    
    return new_image

```

```python
def rescale(image):
    return image / 255.0
```

```python
def normalize_image(image):
    mean = np.mean(image)
    std = np.std(image)
    return (image - mean) / std
```

```python
def convert_to_array(image):
    return np.array(image)
```

```python
def load_and_preprocess(datadir , target_size = (244,244)):
    images = []
    labels = []
    for label in ['yes' , 'no']:
        folder_path = os.path.join(datadir , label)
        for filename in os.listdir(folder_path):
            image_path = os.path.join(folder_path , filename)
            image = Image.open(image_path).convert('RGB')
            image = resize_and_padding(image , target_size)
            image = convert_to_array(image)
            image = rescale(image)
            image = normalize_image(image)
            images.append(image)
            labels.append(1 if label == 'yes' else 0)
    images = np.array(images)
    labels = np.array(labels)
    return images , labels
```

```python
images , labels = load_and_preprocess(datadir)
```

```python
def create_data_generators(train_data, val_data, img_height, img_width, batch_size):
    train_images, train_labels = train_data
    val_images, val_labels = val_data
    
    train_datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    val_datagen = ImageDataGenerator()
    
    train_generator = train_datagen.flow(
        train_images, train_labels, 
        batch_size=batch_size
    )
    
    validation_generator = val_datagen.flow(
        val_images, val_labels, 
        batch_size=batch_size
    )
    
    return train_generator, validation_generator
```

```python
train_images, val_images, train_labels, val_labels = split_data(images, labels)
```

```python

# Create data generators
img_height, img_width = 224, 224
batch_size = 32
train_data = (train_images, train_labels)
val_data = (val_images, val_labels)

train_generator, validation_generator = create_data_generators(
    train_data, val_data, img_height, img_width, batch_size
)
```

```python
def get_generator_length(generator):
    length = 0
    for _ in generator:
        length += 1
    return length
    
# Example usage
train_length = get_generator_length(train_generator)
val_length = get_generator_length(validation_generator)

print(f"Length of train generator: {train_length}")
print(f"Length of validation generator: {val_length}")
```
